```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
# Data
test_types = [
    "Motorettes (AC-GND;\nphase-to-phase)",
    "Motorettes (AC-GND;\nturn-to-turn)",
    "Twisted Pairs\n(AC-GND)",
    "Power Module\n(AC-GND)"
]

electrical_pdiv = [902, 629, 668, 1088]
ae_hits_pdiv    = [1290, 1188, 716, 1004]
ae_spec_pdiv    = [811, 842, 682, 511]

# Bar positions
x = np.arange(len(test_types))
width = 0.25  # bar width

# Custom colors (change these as you like)
colors = {
    "electrical": "tab:blue",
    "ae_hits":    "tab:purple",
    "ae_spec":    "tab:red"
}

# Plot
fig, ax = plt.subplots(figsize=(10, 5))

# Bars (and place number labels on top)
bars1 = ax.bar(x - width, electrical_pdiv, width, label="Electrical PDIV", color=colors["electrical"])
bars2 = ax.bar(x,         ae_hits_pdiv,    width, label="AE hits-derived PDIV", color=colors["ae_hits"])
bars3 = ax.bar(x + width, ae_spec_pdiv,    width, label="AE spectral-derived PDIV", color=colors["ae_spec"])

ax.bar_label(bars1, padding=3)
ax.bar_label(bars2, padding=3)
ax.bar_label(bars3, padding=3)

# Formatting
ax.set_xlabel("Test Type")
ax.set_ylabel("PDIV (V)")
ax.set_xticks(x)
ax.set_xticklabels(test_types)
ax.legend()

ax.set_axisbelow(True)
ax.grid(True, axis="y", linestyle="--", alpha=0.4)

ax.set_title(
    "Comparison of PDIV Measurement Methods",
    fontsize=14,
    pad=12
)

max_value = max(max(electrical_pdiv), max(ae_hits_pdiv), max(ae_spec_pdiv))
ax.set_ylim(0, max_value * 1.2)

plt.tight_layout()
plt.show()
```
